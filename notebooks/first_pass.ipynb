### state_by_state

This is the real bulk of the data - cancer data by US county. 

Seeing as this has 51 files, I will have to read them recursively. 

```python
import os 
import glob 

states_files = glob.glob(os.path.join('../data/capstone_project/state_by_state', "*")) 
```

```python
states_data = {}
for state_file in states_files:
    states_data[state_file] = pd.read_excel(state_file, skiprows=1, sheet_name=None)
```

```python
states_data.__len__()
```

```python
states_data[states_files[0]]['Neoplasms'].head()
```

Now I'd better unite these all into a singular dataset.

```python
state_data_df = states_data[list(states_data.keys())[0]]

for state_data_key in list(states_data.keys())[1:]:
    state_data = states_data[state_data_key]
    for state_cancer_key in state_data.keys():
        state_cancer_data = state_data[state_cancer_key]
        state_data_df[state_cancer_key] = pd.concat([state_data_df[state_cancer_key], state_cancer_data])
```

```python
state_data_df['Neoplasms'].head()
```

Now we have a unified data set. 

With this we can analyze the result.

```python
for sheet in state_data_df.values():
    print(sheet.isna().sum())
```

Now some further analysis of these nulls - which seem to be even across the sheets - is needed.

```python
for sheet in state_data_df.values():
    print(sheet[sheet.isna()['Mortality Rate, 2014*']])
```

As can be seen, these are empty useless rows, so they should be discarded. 

```python
for state_key in state_data_df.keys():
    sheet = state_data_df[state_key]
    state_data_df[state_key] = sheet.drop(sheet[sheet.isna()['Mortality Rate, 2014*']].index)
```

```python
for sheet in state_data_df.values():
    print(sheet.isna().sum())
```
